```python
from sklearn import metrics
import numpy as np
import csv
```

```python
def distance(p1, p2, d):
    sum = 0
    for i in range(d):
        sum = sum + pow(p1[i] - p2[i], 2)
    return sum
```

```python
def my_kmean(data, k,rseed=10):
    rows = len(data)
    dimention = len(data[0])
    clusters = np.zeros(rows)
    rng = np.random.RandomState(rseed)
    i = rng.permutation(data.shape[0])[:k]
    centroids = data[i]
    times = 700
    while True:
        new = np.zeros(rows)
        #making clusters
        for i in range(rows):
            dist = distance(data[i], centroids[0], dimention)
            for j in range(1,k):
                temp = distance(data[i], centroids[j], dimention)
                if temp < dist:
                    dist = temp
                    new[i] = int(j)
        
        # finding new centroids
        total = np.zeros((k,dimention))
        count = np.zeros(k)
          #suming up for each cluster
        for i in range(rows):
            count[int(new[i])] += 1
            total[int(new[i])] = np.add(total[int(new[i])], data[i])
            #dividing with count
        for i in range(k):
            centroids[i] = total[i] / count[i]
        
        
        # checking for changes
        comparison = clusters == new 
        
        if times < 0 or comparison.all():
            #print(centroids)
            return new
        clusters = new
        times -= 1
    return -1
```

```python
np.seterr(divide='ignore', invalid='ignore')

filename = "KMean_Values_final.csv"

with open(filename, 'w') as csvfile: 
    csvwriter = csv.writer(csvfile,delimiter = ',', lineterminator='\r')
    
    for i in range(1,57):
        print(i, end =' ')
        
        fname='C:/Users/Dravyansh/dsa/DM_Project/data/'+str(i)+'.csv'
        
        data=np.genfromtxt(fname,delimiter=',')
        sil = list()
        bol = list()
        
        for j in range(2,11):
            clus = my_kmean(data[:,0:-1],j)
            sil.append(metrics.silhouette_score(data[:,0:-1], clus))
            bol.append(metrics.davies_bouldin_score(data[:,0:-1], clus))
        
        sil.extend(bol)
        
        csvwriter.writerow(sil)
        
        print("done")

print("Completed")
```
